```python
#Importing dependencies
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
```

### Part 1: Prepare the Data

```python
#Read `myopia.csv` into a Pandas DataFrame.
myopia_df = pd.read_csv("Resources/myopia.csv")
myopia_df
```

```python
#Remove the "MYOPIC" column from the dataset
myopia_labels = myopia_df["MYOPIC"]
X = myopia_df.drop(columns=["MYOPIC"])
```

```python
#Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Save scaled dataframe by assigning scaled values
myopic_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
myopic_scaled_df
```

### Part 2: Apply Dimensionality Reduction

```python
#Review Unit 20 Day 1 Activity 7 for similar code
#Initialize PCA model
#preserve 90% of the explained variance in dimensionality reduction
pca = PCA(n_components=0.90)

# Get two principal components for the data.
myopia_pca = pca.fit_transform(myopic_scaled_df)
```

```python
#Transform PCA data to a DataFrame
#Abbreviated Principal Components to "P.C."
myopic_pca_df = pd.DataFrame(
    data=myopia_pca, columns=["P.C. 1", "P.C. 2", "P.C. 3", "P.C. 4", "P.C. 5", "P.C. 6", "P.C. 7", "P.C. 8", "P.C. 9", "P.C. 10"]
)
myopic_pca_df.head()
```

```python
#Fetch the explained variance
pca.explained_variance_ratio_
```

```python
#Initialize t-sne model
tsne = TSNE(learning_rate=150)
#Reducing tsne dimensions
tsne_fit = tsne.fit_transform(myopic_pca_df)
```

```python
#Checking if dimensions were reduced to two with tsne df. shape
tsne_fit.shape
```

```python
#Create a scatter plot of the t-SNE output. Are there distinct clusters?
plt.scatter(tsne_fit[:,0], tsne_fit[:,1])
plt.show()
```

```python
#Add labels to TSNE results to color them and plot; identify possible clusters
plt.scatter(tsne_fit[:,0], tsne_fit[:,1], c=myopia_labels)
plt.show()
```

### Part 3: Perform a Cluster Analysis with K-means

```python
#Use a `for` loop to determine the inertia for each `k` between 1 through 10
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(myopic_scaled_df)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow
```

```python
# Plot the elbow curve to find the best candidate(s) for k
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve for MYOPIC Data')
plt.show()
```

### Part 4: Make a Recommendation

After scaling the data, fitting it to our dataframe, using the T-SNE model, and using k-means to find the elbow curve of our data, I was able to make a few observations about the dataset. Although the plots made using the TSNE model do show a slight definition of 5 possible clusters, they are not very clear despite trying multiple learning rates for it, so it suggests that clustering may not be the best idea. Additionally, the Elbow Curve based off the k-means values displays the elbow of the curve between k=2 to k=3, but closer to k=3. However, the inertia value around the elbow of the curve still remains really high between 6500 and 7000, suggesting clustering may not be the best idea for this dataset. Perhaps if there were more data points, clustering might be more feasible, but the way the dataset is currently wouldn't allow that to provide accurate results.
